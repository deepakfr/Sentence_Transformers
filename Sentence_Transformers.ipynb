```python
model = SentenceTransformer('multi-qa-distilbert-cos-v1')
```

```python
queries = requests.get('https://raw.githubusercontent.com/laxmimerit/machine-learning-dataset/refs/heads/master/text-dataset-for-machine-learning/sbert-queries.txt')
```

```python
queries = queries.text
```

```python
queries = queries.split('\r\n')
```

```python
corpus_embedding = model.encode(corpus, convert_to_tensor=True)
queries_embedding = model.encode(queries, convert_to_tensor=True)
```

```python
len(corpus_embedding[0])
```

# Let normalise vectors for fast *calculations*

```python
normalise_corpus = util.normalize_embeddings(corpus_embedding)
normalise_queries = util.normalize_embeddings(queries_embedding)
```

```python
len(normalise_corpus[0])
```

```python
hits = util.semantic_search(normalise_queries, normalise_corpus,score_function= util.dot_score,top_k=3 )
```

```python
hits
```

```python
print(queries)
```

```python
for query, hit in zip(queries, hits):
  for q_hit in hit:
    id = q_hit['corpus_id']
    score = q_hit['score']
    print(query, corpus[id], score)
```

# K Mean clustering

```python
from sklearn.cluster import KMeans
from sentence_transformers import SentenceTransformer
```

```python
print(len(corpus))
print(len(queries))
```

```python
print(queries)
print(corpus)
```

```python
n_clusters = 5
cluster_model = KMeans(n_clusters=n_clusters)
cluster_model.fit(corpus_embedding)
```

```python
cluster_assignment = cluster_model.labels_
print(cluster_assignment)
```

```python
cluster_sentences = [[] for i in range(n_clusters)]
```

```python
cluster_sentences
```

```python
for sentence_id, cluster_id in enumerate(cluster_assignment):
  cluster_sentences[cluster_id].append(corpus[sentence_id])
```

```python
cluster_sentences
```

```python
for i, cluster in enumerate(cluster_sentences):
  print("\nCluster ", i+1)
  print(cluster)
  print()
```

Fast Clustering

Fast Clustering is used for larger datasets. it can tune 50k sentences and cluster it

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import time
```

```python
df =pd.read_csv('quora_duplicate_questions.tsv', sep='\t')
```

```python
df
```

```python
sentences = df['question1'].to_list()[:1000]
```

```python
len(sentences)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences_embedding = model.encode(sentences, convert_to_tensor=True, batch_size=64, show_progress_bar=True)
```

```python
clusters = util.community_detection(sentences_embedding, min_community_size=5, threshold=0.5)
```

```python
clusters
```

```python
for i, cluster in enumerate(clusters):
  print("\nCluster #{}, #{} Questions".format(i+1, len(cluster)))
  for sentence_id in cluster[0:3]:
    print("\t", sentences[sentence_id])
    print("\t", "...")
```
